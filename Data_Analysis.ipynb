```python
import pandas as pd
import numpy as np
import dataframe_image as dfi

from tabulate import tabulate

def compute_corr(data, Param):
    #corr_v_ozone = data[Param].corr(data['omean'])
    #corr_v_pm2 = data[Param].corr(data['pmean'])
    corr_v_aqi = data[Param].corr(data['aqi'])
    return (corr_v_aqi)


def compute_corr_all(filename):
    data = pd.read_csv(filename, index_col=False)
    
    corr_df = pd.DataFrame({'Parameter': pd.Series(dtype='str'),'corr_v_aqi':pd.Series(dtype='float'),
                            })
    #'corr_v_ozone': pd.Series(dtype='float'), 'corr_v_pm2': pd.Series(dtype='float')
    
    (corr_v_aqi) = compute_corr(data=data, Param="temp")
    new_df = pd.DataFrame([['Temp', corr_v_aqi]], 
                          columns=['Parameter', 'corr_v_aqi'])
    corr_df = pd.concat([corr_df, new_df])
    
    (corr_v_aqi) = compute_corr(data=data, Param="precip")
    new_df = pd.DataFrame([['precip', corr_v_aqi]], 
                          columns=['Parameter', 'corr_v_aqi'])
    corr_df = pd.concat([corr_df, new_df])
    
    (corr_v_aqi) = compute_corr(data=data, Param="windspeed")
    new_df = pd.DataFrame([['windspeed', corr_v_aqi]], 
                          columns=['Parameter', 'corr_v_aqi'])
    corr_df = pd.concat([corr_df, new_df])
    
    (corr_v_aqi) = compute_corr(data=data, Param="humidity")
    new_df = pd.DataFrame([['humidity', corr_v_aqi]], 
                          columns=['Parameter', 'corr_v_aqi'])
    corr_df = pd.concat([corr_df, new_df])
    
#     (corr_v_aqi, corr_v_ozone, corr_v_pm2) = compute_corr(data=data, Param="omean")
#     new_df = pd.DataFrame([['omean', corr_v_aqi, corr_v_ozone, corr_v_pm2]], 
#                           columns=['Parameter', 'corr_v_aqi', 'corr_v_ozone', 'corr_v_pm2'])
#     corr_df = pd.concat([corr_df, new_df])
    
#     (corr_v_aqi, corr_v_ozone, corr_v_pm2) = compute_corr(data=data, Param="pmean")
#     new_df = pd.DataFrame([['pmean', corr_v_aqi, corr_v_ozone, corr_v_pm2]], 
#                           columns=['Parameter', 'corr_v_aqi', 'corr_v_ozone', 'corr_v_pm2'])
#     corr_df = pd.concat([corr_df, new_df])
    
    return corr_df

df_corr = compute_corr_all(filename= "../Data_Prep_LA/WeatherAndPollution_2011_2020.csv")
df_corr = df_corr.reset_index()
df_corr = df_corr.drop(['index'], axis=1)
dfi.export(df_corr,"../CorrTables/LACorrTable.png")

df_corr = compute_corr_all(filename= "../Data_Prep_Baker/WeatherAndPollution_2011_2020.csv")
df_corr = df_corr.reset_index()
df_corr = df_corr.drop(['index'], axis=1)
dfi.export(df_corr,"../CorrTables/BakerCorrTable.png")

df_corr = compute_corr_all(filename= "../Data_Prep_AZ/WeatherAndPollution_2011_2020.csv")
df_corr = df_corr.reset_index()
df_corr = df_corr.drop(['index'], axis=1)
dfi.export(df_corr,"../CorrTables/AZCorrTable.png")

df_corr = compute_corr_all(filename= "../Data_Prep_TX/WeatherAndPollution_2011_2020.csv")
df_corr = df_corr.reset_index()
df_corr = df_corr.drop(['index'], axis=1)
dfi.export(df_corr,"../CorrTables/TXCorrTable.png")

# df_corr = compute_corr_all(filename= "../Data_Prep_ALL/WeatherAndPollution_2011_2020.csv")
# df_corr = df_corr.reset_index()
# df_corr = df_corr.drop(['index'], axis=1)
# dfi.export(df_corr,"../CorrTables/ALLCorrTable.png")

#print(tabulate(df_corr, headers = 'keys', tablefmt = 'fancy_grid'))

#df_styled = df_corr.style.background_gradient()

print('done')
```

```python
import matplotlib.pyplot as plt

def create_save_plt(df, param1, param2, pre):
    plt.figure(1)
    plt.scatter(df[param1], df[param2])
    
    plt.xlabel(param1)
    plt.ylabel(param2)
    plt.ylim(0, 400)
    
    filename= "../CorrTables/"+pre+"_"+param1+"_v_" + param2+".png"
    plt.savefig(filename)
    plt.close()

def create_plot_all(filename, pre):
    df = pd.read_csv(filename)
    
    create_save_plt(df, 'temp', 'aqi', pre)
    create_save_plt(df, "humidity", "aqi", pre)
    create_save_plt(df, "precip", 'aqi', pre)
    create_save_plt(df, 'windspeed', 'aqi', pre)
    
create_plot_all(filename='../Data_Prep_Baker/WeatherAndPollution_2011_2020.csv', pre='Baker')
create_plot_all(filename='../Data_Prep_LA/WeatherAndPollution_2011_2020.csv', pre='LA')
create_plot_all(filename='../Data_Prep_AZ/WeatherAndPollution_2011_2020.csv', pre='AZ')
create_plot_all(filename='../Data_Prep_TX/WeatherAndPollution_2011_2020.csv', pre='TX')
#create_plot_all(filename='../Data_Prep_ALL/WeatherAndPollution_2011_2020.csv', pre='ALL')

print('done')
```

```python
import pandas as pd
import matplotlib.pyplot as plt


def histo(pre):
    df = pd.read_csv('../Data_Prep_'+pre+'/WeatherAndPollution_2011_2020.csv')
    
    # Set the figure size
    plt.rcParams["figure.figsize"] = [7.00, 3.50]
    plt.rcParams["figure.autolayout"] = True
    plt.xlabel('AQI')
    plt.ylabel('Frequency')
    plt.axis([0,200, 0, 2000])
    plt.title('Histogram of AQI')


    # Data points for the histogram
    k = df['aqi']

    # Plot the histogram
    plt.hist(k)

    # Save the histogram
    plt.savefig('../ExtraData/'+pre+'hist.png')

    # Display the plot
    plt.show()
    

histo("SF")
histo("FL")
histo("NY")
```
